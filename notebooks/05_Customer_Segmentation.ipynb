Customer Segmentation Strategy

The customer base is segmented using the RFM framework.

Segments include:

- Champions
- Loyal Customers
- Potential Loyalists
- New Customers
- Promising
- Need Attention
- At Risk
- Can't Lose Them
- Hibernating
- Lost

The objective is to enable targeted marketing strategies based on customer behavior and value.

```python
import pandas as pd
```

```python
customer_rfm = pd.read_csv('../data/processed/customer_rfm.csv')
```

```python
customer_rfm['Segment'] = 'Others'
```

```python
champion_mask = (
    (customer_rfm['R_Score'] >= 4) &
    (customer_rfm['F_Score'] >= 4) &
    (customer_rfm['M_Score'] >= 4)
)


customer_rfm.loc[champion_mask, 'Segment'] = 'Champions'
```

```python
customer_rfm['Segment'].value_counts()
```

Champions 

Objective

Identify the most valuable customers who have purchased recently, buy frequently, and generate the highest revenue.

Finding

A total of 856 customers were classified as Champions.

These customers achieved high scores across all three RFM dimensions:

High Recency → They purchased recently.
High Frequency → They purchase frequently.
High Monetary Value → They spend significantly more than other customers.
Business Insight

Champions represent the company's most valuable customer segment. They contribute substantially to overall revenue while also demonstrating strong engagement and purchase loyalty. Retaining these customers is generally more cost-effective than acquiring new ones.

Business Recommendation
- Provide VIP or premium membership benefits.
- Offer early access to new products and exclusive promotions.
- Deliver personalized recommendations and premium customer support.
- Use this segment for referral and ambassador programs to attract new customers.

```python
loyal_mask =(
    (customer_rfm['Segment'] == 'Others') &
    (customer_rfm['F_Score'] >= 4) &
    (customer_rfm['R_Score'] >= 3)
)

customer_rfm.loc[loyal_mask, 'Segment'] = 'Loyal Customer'
```

```python
customer_rfm['Segment'].value_counts()
```

Loyal Customers

Objective

Identify customers who purchase frequently and remain actively engaged with the business.

Finding

751 customers were classified as Loyal Customers.

These customers exhibit high purchase frequency and relatively recent activity, indicating consistent engagement with the platform.

Business Insight

Loyal Customers represent a stable source of recurring revenue. While they may not always be the highest spenders, their consistent purchasing behavior makes them valuable for long-term customer retention.

Business Recommendation

- Reward loyalty through membership programs.
- Provide exclusive discounts and personalized offers.
- Encourage progression toward the Champions segment through targeted campaigns.

```python
potential_loyalists_mask =(
    (customer_rfm['Segment'] == 'Others') &
    (customer_rfm['F_Score'].isin([2,3])) &
    (customer_rfm['R_Score'] >= 4)
)

customer_rfm.loc[potential_loyalists_mask, 'Segment'] = 'Potential Loyalists'
```

```python
customer_rfm['Segment'].value_counts()
```

Potential Loyalists

Objective

Identify recently acquired customers who have started engaging with the business and show strong potential to become loyal customers.

Finding

A total of 652 customers were classified as Potential Loyalists.

These customers have made recent purchases but have not yet reached a high purchase frequency, indicating that they are still in the early stages of their customer journey.

Business Insight

Potential Loyalists represent a valuable growth opportunity. With appropriate engagement strategies, many of these customers can transition into the Loyal Customer or Champion segments, increasing their lifetime value.

Business Recommendation
- Send personalized follow-up emails after purchases.
- Offer limited-time discounts to encourage repeat purchases.
- Recommend related products based on previous orders.
- Enroll customers in loyalty or rewards programs early to strengthen long-term engagement.

```python
new_customer_mask =(
    (customer_rfm['Segment'] == 'Others') &
    (customer_rfm['F_Score'] == 1) &
    (customer_rfm['R_Score'] >= 4)
)

customer_rfm.loc[new_customer_mask, 'Segment'] = 'New Customer'
```

```python
customer_rfm['Segment'].value_counts()
```

New Customers

Objective

Identify customers who have recently made their first purchase and have just entered the customer lifecycle.

Finding

A total of 182 customers were identified as New Customers.

These customers have purchased very recently but have not yet established a repeat purchasing pattern.

Business Insight

New Customers represent the first stage of the customer journey. Their future value depends on whether the business can successfully encourage additional purchases and build long-term engagement.

Business Recommendation

- Send a welcome email after the first purchase.
- Offer a discount for the second purchase.
- Recommend complementary products.
- Provide an excellent onboarding and post-purchase experience.

```python
cant_lose_mask = (
    (customer_rfm['Segment'] == 'Others') &
    (customer_rfm['R_Score'] <= 2) &
    (customer_rfm['F_Score'] >= 4) &
    (customer_rfm['M_Score'] >= 4)
)

customer_rfm.loc[ cant_lose_mask, 'Segment'] = 'Cannot Lose Them'
```

```python
customer_rfm['Segment'].value_counts()
```

```python
at_risk_mask =(
    (customer_rfm['Segment'] == 'Others') &
    (customer_rfm['F_Score'] >= 3) &
    (customer_rfm['R_Score'] <= 2)
)

customer_rfm.loc[at_risk_mask, 'Segment'] = 'At Risk'
```

```python
customer_rfm['Segment'].value_counts()
```

⚠️ At Risk

Objective

Identify customers who were previously active purchasers but have not made a recent purchase and are at risk of churning.

Finding

A total of 778 customers were classified as At Risk.

These customers have a history of frequent purchases but have shown low recent activity, indicating a high probability of customer churn if no action is taken.

Business Insight

At Risk customers have already demonstrated strong engagement in the past, making them more valuable than newly acquired customers. Re-engaging these customers is often more cost-effective than acquiring new ones.

Business Recommendation

- Launch win-back campaigns with personalized offers.
- Send reminder emails highlighting new products or promotions.
- Provide limited-time discounts to encourage reactivation.
- Prioritize high-value At Risk customers for personalized outreach.

```python
hibernating_mask =(
    (customer_rfm['Segment'] == 'Others') &
    (customer_rfm['F_Score'] <= 2) &
    (customer_rfm['R_Score'] == 2)
)

customer_rfm.loc[hibernating_mask, 'Segment'] = 'Hibernating'
```

```python
customer_rfm['Segment'].value_counts()
```

```python
customer_rfm[
    (customer_rfm['R_Score'] <= 2) &
    (customer_rfm['F_Score'] <= 2)
].shape
```

```python
customer_rfm.groupby('Segment')[['Recency', 'Frequency', 'Monetary']].mean().round(1)
```

```python
lost_mask = (
    (customer_rfm['Segment'] == 'Others') &
    (customer_rfm['R_Score'] == 1) &
    (customer_rfm['F_Score'] <= 2)
)

customer_rfm.loc[lost_mask, 'Segment'] = 'Lost'
```

```python
customer_rfm['Segment'].value_counts()
```

```python
customer_rfm['R_Score'].value_counts().sort_index()
```

```python
pd.crosstab(
    customer_rfm['R_Score'],
    customer_rfm['Segment']
)
```

### 📌 Customer Segmentation Summary

In this section, customers were segmented using the **RFM (Recency, Frequency, Monetary)** framework. Rather than relying solely on predefined templates, the segmentation rules were refined to better reflect the customer lifecycle and business objectives.

The final customer segments are:

* **Champions:** Recent, frequent, and high-value customers. They represent the company's most valuable customer group.
* **Loyal Customers:** Customers who purchase regularly and maintain a strong relationship with the business.
* **Potential Loyalists:** Customers who have shown promising purchasing behavior and have the potential to become loyal customers with appropriate engagement.
* **New Customers:** Recently acquired customers with limited purchase history who require nurturing.
* **Cannot Lose Them:** High-value customers who have become inactive and require immediate retention efforts.
* **At Risk:** Customers who previously engaged with the business but have recently reduced their activity and may churn.
* **Hibernating:** Customers with low recent activity who still have potential to be reactivated through targeted marketing campaigns.
* **Lost:** Customers who have not purchased for a long period and exhibit very low engagement, indicating a low probability of returning.

This segmentation provides a solid foundation for customer analysis and enables the development of personalized marketing strategies, customer retention campaigns, and business decision-making based on customer value and behavior.
